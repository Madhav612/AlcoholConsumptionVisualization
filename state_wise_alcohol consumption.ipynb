```python
fig,axes = plt.subplots(ncols=3,figsize=(15,5))
sns.boxplot(y=df['Cost per capita ($)'],ax=axes[0]).set_title('Alcohol consumption Cost per Capita in $')
sns.swarmplot(y=df['Cost per capita ($)'],ax=axes[0],color='red')
sns.boxplot(y=df['Cost per drink ($)'],ax=axes[1]).set_title('Alcohol cost per drink in $')
sns.swarmplot(y=df['Cost per drink ($)'],ax=axes[1],color='red')
sns.boxplot(y=df['Total Cost ($)'],ax=axes[2]).set_title('Total cost to states for alcohol consumption in Billions')
sns.swarmplot(y=df['Total Cost ($)'],ax=axes[2],color='red')
```

This boxplot represents the cost of alcohol consumption for each state. From the first barplot, we can analyze that median amount that states spend behind excessive drinking is around 800 dollars. Median of alcohol price per drink is around 2 dollars and median expense for states is around 4 billion dollars per year which is substantially large.  

```python
fig = px.choropleth(df,
                    locations='code',
                    color='Total Cost ($)',
                    color_continuous_scale='spectral_r',
                    hover_name='Location',
                    locationmode='USA-states',
                    labels={'Total Cost ($)':'Total Cost to States in Billions'},
                   scope='usa')
fig.add_scattergeo(
        locations=df['code'],
    locationmode = 'USA-states',
    text=df['code'],
    mode='text'
)
fig.update_layout(title={'text':'Total Cost of Alcohol consumption by state',
                         'xanchor':'center',
                        'yanchor':'top',
                        'x':0.5})
fig.show()
```

As we can see from the choropleth map above, california spends substanially more because of alcohol consumption compared to other states. But this is not the accurate measure to calcualte the alcohol consumption as states have different population and states with more population has higher overall consumption by states.
So let's try to plot the total cost to states per capita as this would be an accurate measure to compare all states

```python
fig = px.choropleth(df,
                    locations='code',
                    color='Cost per capita ($)',
                    color_continuous_scale='spectral_r',
                    hover_name='Location',
                    locationmode='USA-states',
                    labels={'Cost per capita ($)':'Total Cost per capita in $'},
                   scope='usa')
fig.add_scattergeo(
        locations=df['code'],
    locationmode = 'USA-states',
    text=df['code'],
    mode='text'
)
fig.update_layout(title={'text':'Total Cost of Alcohol consumption by state per capita',
                         'xanchor':'center',
                        'yanchor':'top',
                        'x':0.5})
fig.show()
```

Visualizing the choropleth is quite difficult as some states in east america are in small porximity. So let's plot using scatter_geo() instead where we can visualize the graph more accurately.

```python
fig = px.scatter_geo(df, locations="code", color="Cost per capita ($)",
                     hover_name="Location", size="Cost per capita ($)",
                     projection="albers usa",locationmode='USA-states',color_continuous_scale='spectral_r',text='code',
                    title='Total Cost($) of Alcohol consumption by state per capita',width=1000,height=800)
fig.show()
```
